```python
#1.
import seaborn as sns
import plotly.graph_objects as go
titanic=sns.load_dataset('titanic')
```

```python
titanic
```

```python
#1.
fig= go.Figure()
fig.add_trace(go.Scatter(x=titanic.age, y=titanic.fare,mode='markers'))

fig.show()
```

```python
#2.
import plotly.express as px


data = px.data.tips()

figure = px.box(data_frame=data, x="time", y="tip")

figure.show()
```

```python
#3.
fig = go.Figure(data = [go.Histogram(x = data.sex,y=data.total_bill)])
fig.show()
```

```python
#3.
import plotly.express as px

data = px.data.tips()

figure = px.histogram(data_frame=data, x="sex", y="total_bill", color="day", pattern_shape="smoker")

figure.show()
```

```python
#4.
import plotly.express as px

# Load the tips dataset from Plotly Express
data = px.data.iris()
fig= px.scatter_matrix(data_frame= data, dimensions=["sepal_length", "sepal_width", "petal_length", "petal_width"],color='species')
fig.show()
```

```python
#5.
#A distplot is a plot that displays a combination of statistical representations of numerical data, such as a histogram, kernel density estimation or normal curve, and rug plot.
import plotly.express as px

data = px.data.tips()

figure = px.histogram(data_frame=data, x="total_bill", y="tip", color="sex", marginal="rug")

figure.show()
```
